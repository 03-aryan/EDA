```python
car = cardf.drop(['car_ID', 'symboling', 'CarName', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight', 'boreratio', 'stroke','citympg', 'highwaympg',],axis=1)
```

```python
car.head()
```

```python
core = car.corr()
```

### Heat map representing how different features correlate with each other we tend to use heat map for feature selection to make ML models.

```python
sns.heatmap(core,xticklabels=core.columns,yticklabels=core.columns,annot =True)
```

```python
sns.heatmap(core,cmap='RdBu',xticklabels=core.columns,yticklabels=core.columns,annot =True)
```

### Pairplot tells us relationship between the 2 variables or features in the data set and how its related to ever other variable in data set. 

```python
sns.pairplot(car)
```

### Relational plot shows relation between two particular features using help us to differentiate how both feature coincide with eachother 

```python
sns.relplot(x='horsepower',y='enginesize',hue='fueltype', data=car)
```

```python
sns.relplot(x='horsepower',y='enginesize', data=car)
```

### Histogram: It basically represents the range of values the longer the height of bar more number of values fall into that particular section.

```python
sns.distplot(car['peakrpm'])
```

### In box plot the mid black line inside the box represents median value of particular feature, the box plot also represents maximum as well as minimum value of the feature, the diamond black at the top is an outlier.

```python
sns.catplot(y='peakrpm',kind='box',data=car)
```

### Violin plot is basically a mix of box plot and kernel density plot, here the plot represents the variation in peakrpm in the second graph we compared diesel and gas fuel type with reference to horsepower providing gas vehicles have more horsepower.

```python
sns.violinplot(y='peakrpm',data=car)
```

```python
sns.violinplot(x='fueltype',y='horsepower',data=car)
```

### Here in joint plot in kernel density format we show how two features are related and where the density of those features are highest.

```python
sns.jointplot(x='horsepower',y='peakrpm',data=car,kind='kde')
```

```python
sns.jointplot(x='horsepower',y='peakrpm',hue='drivewheel',data=car,kind='kde')
```
